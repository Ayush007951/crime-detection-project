## Step 6: Year-wise Crime Trend Analysis

This step analyzes how the total number of crime cases has changed over the years.
It helps in identifying increasing or decreasing crime patterns over time.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/crime_india.csv")
df.head()
```

```python
df['Cases_Property_Stolen'] = pd.to_numeric(
    df['Cases_Property_Stolen'], errors='coerce'
)

yearly_crime = df.groupby('Year')['Cases_Property_Stolen'].sum()

plt.figure()
yearly_crime.plot()
plt.xlabel('Year')
plt.ylabel('Total Property Stolen Cases')
plt.title('Year-wise Crime Trend')
plt.show()
```

## Step 7: Property Stolen vs Property Recovered Analysis

This step compares the number of cases in which property was stolen
with the number of cases in which property was successfully recovered.


```python
# convert both columns to numeric
df['Cases_Property_Recovered'] = pd.to_numeric(
    df['Cases_Property_Recovered'], errors='coerce'
)

df['Cases_Property_Stolen'] = pd.to_numeric(
    df['Cases_Property_Stolen'], errors='coerce'
)

# calculate totals
total_stolen = df['Cases_Property_Stolen'].sum()
total_recovered = df['Cases_Property_Recovered'].sum()

# bar chart comparison
plt.figure()
plt.bar(['Stolen', 'Recovered'], [total_stolen, total_recovered])
plt.xlabel('Category')
plt.ylabel('Number of Cases')
plt.title('Property Stolen vs Property Recovered')
plt.show()
```

## Step 8: Crime Hotspot Identification

This step identifies high-crime areas (hotspots) by analyzing
the total number of property stolen cases across different regions.


```python
import pandas as pd
import matplotlib.pyplot as plt

# reload dataset cleanly
df = pd.read_csv("data/crime_india.csv")

# rename columns (VERY IMPORTANT)
df = df.rename(columns={
    'Area_Name': 'Area',
    'Group_Name': 'Crime_Group',
    'Sub_Group_Name': 'Crime_Type'
})

# convert stolen cases to numeric
df['Cases_Property_Stolen'] = pd.to_numeric(
    df['Cases_Property_Stolen'], errors='coerce'
)

# area-wise total stolen cases
area_crime = df.groupby('Area')['Cases_Property_Stolen'].sum()

# top 10 crime hotspots
top_hotspots = area_crime.sort_values(ascending=False).head(10)

# plot
plt.figure()
top_hotspots.plot(kind='bar')
plt.xlabel("Area")
plt.ylabel("Total Property Stolen Cases")
plt.title("Top 10 Crime Hotspot Areas")
plt.show()
```

```python
# crime type distribution
crime_type_counts = df['Crime_Type'].value_counts().head(10)

plt.figure()
crime_type_counts.plot(kind='bar')
plt.xlabel("Crime Type")
plt.ylabel("Number of Cases")
plt.title("Top Crime Types (Property Crimes)")
plt.show()
```

## Step 9: Crime Hotspot Prediction using K-Means Clustering

In this step, an unsupervised machine learning model (K-Means Clustering)
is used to group areas into different crime risk levels based on the
number of property stolen cases. This helps in predicting crime hotspots.


```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# prepare area-wise crime data
model_data = df.groupby('Area')[['Cases_Property_Stolen']].sum()
model_data = model_data.dropna()

# scaling the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(model_data)
```

```python
# apply K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
model_data['Crime_Risk_Level'] = kmeans.fit_predict(scaled_data)

model_data.head()
```

```python
plt.figure()
plt.scatter(
    model_data.index,
    model_data['Cases_Property_Stolen'],
    c=model_data['Crime_Risk_Level']
)
plt.xticks(rotation=90)
plt.xlabel("Area")
plt.ylabel("Total Property Stolen Cases")
plt.title("Crime Hotspot Prediction using K-Means Clustering")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(
    area_crime['Cases_Property_Stolen'],
    area_crime['Crime_Risk_Level']
)

plt.xlabel("Total Property Stolen Cases")
plt.ylabel("Crime Risk Cluster")
plt.title("Crime Hotspot Clustering using K-Means")

plt.show()
```
